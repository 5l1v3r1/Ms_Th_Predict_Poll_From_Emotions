```python
#######################################################################################
#### 4_Descriptive statistics
#######################################################################################
### This code allows to generate the plots for the descriptive statistics
#######################################################################################

### RUN WITH PYTHON 3 ###
### Just change the directory for loading the file at the beginning of code ###

import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
db = np.load ('C:/Users/Victor/images/resfinal.npy') #here change directory
```

```python
label = ['SADNESS','NEUTRAL','CONTEMPT','DISGUST','ANGER','SURPRISE','FEAR','HAPPINESS']
```

```python
#generate the graph for all emotions in the database
def graph_emo(db_emo):

    List = [0]*8
        
    for line in db_emo[1:,4:12]:
        for position, item in enumerate(line):
            List[position] += float(item)
    
    Tot = sum(List)
    
    N= 8
    ind = np.arange(N)  # the x locations for the groups
    width = 0.35       # the width of the bars
    
    plt.subplot(1, 2, 1)
    plt.barh(ind, List, width, color=plt.cm.Set1(np.arange(8)/8.))
    plt.yticks(list(range(0,8)), label, rotation='horizontal')
    plt.xlim( 0, max(List)+150*1.2 ) 
    for i, v in enumerate(List):
        plt.text(v +10, i -0.15 , str(np.round(100*v/Tot,decimals=1)) +'%', color='blue', fontsize = 14)
    fig = plt.gcf()
    fig.set_size_inches(14,6)
       
    plt.subplot(1, 2, 2)
    plt.pie(List, labels=None,labeldistance =1.15, autopct='%1.1f%%', shadow=False, startangle=90, colors = plt.cm.Set1(np.arange(8)/8.))
    plt.axis('equal')
    plt.legend(label,  loc="upper left", bbox_to_anchor=(0, 1))
    plt.title('ALL EMOTIONS',y=-0.05, fontsize =20)
    plt.show()
    
```

```python
graph_emo(db)
```

```python
#generate the graph for main emotions in the database
def graph_emo_main(db_emo):

    List = [0]*8
        
    for line in db_emo[1:,12:]:
        List[int(line[0])] += float(line[1])
    
    Tot = sum(List)
    
    N= 8
    ind = np.arange(N)  # the x locations for the groups
    width = 0.35       # the width of the bars
    
    plt.subplot(1, 2, 1)
    plt.barh(ind, List, width, color=plt.cm.Set1(np.arange(8)/8.))
    plt.yticks(list(range(0,8)), label, rotation='horizontal')
    plt.xlim( 0, max(List)+150*1.2 ) 
    for i, v in enumerate(List):
        plt.text(v +10, i -0.15 , str(np.round(100*v/Tot,decimals=1)) +'%', color='blue',fontsize = 14)
    fig = plt.gcf()
    fig.set_size_inches(14,6)
       
    plt.subplot(1, 2, 2)
    plt.pie(List, labeldistance =1.15, autopct='%1.1f%%', shadow=False, startangle=90, colors = plt.cm.Set1(np.arange(8)/8.))
    plt.axis('equal')
    plt.legend(label,  loc="upper left", bbox_to_anchor=(0, 1))
    plt.title('MAIN EMOTIONS',y=-0.05,fontsize = 20)
    plt.show()
```

```python
graph_emo_main(db)
```

```python
dbp2 = pd.DataFrame(data=db[1:,1:],    # values
...              index=db[1:,14],    # 14th column as index, ie the sign of the poll trend
...              columns=db[0,1:])
```

```python
#plot according to the poll trends : upward trend, main emotions
List = [0]*8

for line in dbp2.loc[[b'1',b'0']].values[1:,11:]:
        List[int(line[0])] += float(line[1])


Tot = sum(List)

N= 8
ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

plt.subplot(1, 2, 1)
plt.barh(ind, List, width, color=plt.cm.Set1(np.arange(8)/8.))
plt.yticks(list(range(0,8)), label, rotation='horizontal')
plt.xlim( 0, max(List)+150*1.2 ) 
for i, v in enumerate(List):
    plt.text(v +10, i -0.15 , str(np.round(100*v/Tot,decimals=1)) +'%', color='blue', fontsize = 14)
fig = plt.gcf()
fig.set_size_inches(14,6)

plt.subplot(1, 2, 2)
plt.pie(List, labeldistance =1.15, autopct='%1.1f%%', shadow=False, startangle=90, colors = plt.cm.Set1(np.arange(8)/8.))
plt.axis('equal')
plt.legend(label,  loc="upper left", bbox_to_anchor=(0, 1))
plt.title('UPWARD TREND',y=-0.05, fontsize=20)
plt.show()
```
